```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cov = 1/4 * np.array([[5,3],[3,5]]) # det(cov) = 1 and has eigenvalues 2 and 1/2
data = np.random.multivariate_normal([0,0], cov, 1000)
plt.plot(data[:,0], data[:,1], 'bo')
```
